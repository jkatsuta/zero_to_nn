```python
def sigmoid(x):
    return 1. / (1 + np.exp(-x))

def relu(x):
    return np.maximum(x, 0.)

def init_w(i, node_num):
    if i == 1:
        v = np.random.randn(node_num, node_num) * 1.0
    elif i == 2:
        v = np.random.randn(node_num, node_num) * 0.01
    elif i == 3:
        v = np.random.randn(node_num, node_num) * np.sqrt(1. / node_num)
    elif i == 4:
        v = np.random.randn(node_num, node_num) * np.sqrt(2. / node_num)
    return v
        
def sim_nn(i_init_w, f='sigmoid'):
    x = np.random.randn(1000, 100)
    node_num = 100
    n_hidden_layer = 5
    activations = {}

    for i in range(n_hidden_layer):
        if i != 0:
            x = activations[i-1]
        w = init_w(i_init_w, node_num)
        z = np.dot(x, w)
        activations[i] = eval(f)(z)
    return activations
```

```python
%matplotlib inline

for i in range(4):
    activations = sim_nn(i + 1)

    fig, axes = plt.subplots(1, len(activations), figsize=(12, 3), sharey=True)
    plt.subplots_adjust(wspace=0.05)
    for i, a in activations.items():
        axes[i].hist(a.flatten(), bins=np.linspace(0, 1, 30))
        axes[i].set_title('Layer ' + str(i+1))
```

```python
%matplotlib inline

for i in range(4):
    activations = sim_nn(i + 1, 'relu')

    fig, axes = plt.subplots(1, len(activations), figsize=(12, 3), sharey=True)
    plt.subplots_adjust(wspace=0.05)
    for i, a in activations.items():
        axes[i].hist(a.flatten(), bins=np.linspace(0, 3, 30))
        axes[i].set_title('Layer ' + str(i+1))
        axes[i].set_yscale('log')
        axes[i].set_ylim([1, axes[i].get_ylim()[1]])
```

```python
from ch05.twolayers import TwoLayerNet
import common.optimizer as opt

size_hidden = 1000
size_batch = 100
max_epoch = 200

network = TwoLayerNet(x_train.shape[1], size_hidden, y_train.shape[1])
optimizer = opt.SGD(lr=0.01)

size_train = x_train.shape[0]
iter_per_epoch = max(size_train / size_batch, 1)

list_acc_train, list_acc_test = [], []
i, n_epoch = 0, 0
while 1:
    batch_mask = np.random.choice(size_train, size_batch)
    x_batch = x_train[batch_mask]
    y_batch = y_train[batch_mask]
    
    grads = network.gradient(x_batch, y_batch)
    optimizer.update(network.params, grads)
    i += 1
    
    if i % iter_per_epoch == 0:
        acc_train = network.accuracy(x_train, y_train)
        acc_test = network.accuracy(x_test, y_test)
        list_acc_train.append(acc_train)
        list_acc_test.append(acc_test)
        n_epoch += 1
        
        if n_epoch > max_epoch:
            break
```
